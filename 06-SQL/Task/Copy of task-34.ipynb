Some feature details of the dataset:

| Feature | Description |
| :--: | :-- |
| A | Electoral Process |
| B | Political Pluralism and Participation |
| C | Functioning of Government |
| D | Freedom of Expression and Belief |
| E | Associational and Organizational Rights |
| F | Rule of Law |
| G | Personal Autonomy and Individual Rights |
| CL | Civil Liberties Scores |
| Status | F=Free, PF=Partly Free, NF=Not Free |


###`Q-1` Find out top 10 countries' which have maximum A and D values.


###`Q-2` Find out highest CL value for 2020 for every region. Also sort the result in descending order. Also display the CL values in descending order.

```python
# 1. Find out top 10 countries' which have maximum A and D values.
SELECT t1.Country,MAX(t1.A)AS 'MAX_A', MAX(t2.D) AS 'MAX_D'
FROM demodatabase.country_ab t1
JOIN demodatabase.country_cd t2
ON t1.Country = t2.Country
GROUP BY t1.Country
ORDER BY MAX_A DESC ,MAX_D DESC LIMIT 10

# 2.Find out highest CL value for 2020 for every region. Also sort the result in descending order. Also display the CL values in descending order.
SELECT t2.Region , MAX(t1.CL) AS 'MAX_CL'
FROM demodatabase.country_cl t1
JOIN demodatabase.country_efg t2
ON t1.Country = t2.Country
WHERE t1.Edition = 2020
GROUP BY t2.Region
ORDER BY MAX_CL DESC

```

###`Q-3` Find top-5 most sold products.

### `Q-4:` Find sales man who sold most no of products.

### `Q-5:` Sales man name who has most no of unique customer.

###`Q-6:` Sales man who has generated most revenue. Show top 5.

###`Q-7:` List all customers who have made more than 10 purchases.

### `Q-8` : List all salespeople who have made sales to more than 5 customers.

### `Q-9:` List all pairs of customers who have made purchases with the same salesperson.



```python
# 3.Find top-5 most sold products.
SELECT t1.Name,SUM(t2.Quantity) AS 'SUM_q'
FROM demodatabase.products t1
JOIN demodatabase.sales1 t2
ON t1.ProductID = t2.ProductID
GROUP BY t1.Name
ORDER BY SUM_q DESC LIMIT 5

# 4. Find sales man who sold most no of products.
SELECT SalesPersonID , SUM(Quantity) AS 'SUM_q'
FROM demodatabase.sales1
GROUP BY SalesPersonID
ORDER BY SUM_q DESC LIMIT 1

# 5.Sales man name who has most no of unique customer.
SELECT t1.SalesPersonID , COUNT(DISTINCT CustomerID) AS 'UNIQUE_CUST'
FROM demodatabase.sales1 t1
GROUP BY t1.SalesPersonID
ORDER BY UNIQUE_CUST DESC


# 6. Sales man who has generated most revenue. Show top 5.
SELECT t1.SalesPersonID, ROUND(SUM(t1.Quantity * t2.Price),2) AS 'Revenue'
FROM demodatabase.sales1 t1
JOIN demodatabase.products t2
ON t1.ProductID  = t2.ProductID
JOIN demodatabase.employees t3
ON t1.SalesPersonID = t3.EmployeeID
GROUP BY t1.SalesPersonID
ORDER BY Revenue DESC LIMIT 5

# 7.List all customers who have made more than 10 purchases.
SELECT CustomerID , COUNT(SalesID) AS 'Num_Of_Purchases'
FROM demodatabase.sales1
GROUP BY CustomerID
HAVING Num_Of_Purchases > 10
ORDER BY Num_Of_Purchases DESC

# 8. List all salespeople who have made sales to more than 5 customers.

SELECT SalesPersonID, COUNT(DISTINCT CustomerID) AS 'Cust_Associated'
FROM demodatabase.sales1
GROUP BY SalesPersonID
HAVING Cust_Associated > 5
ORDER BY Cust_Associated DESC;

# 9.List all pairs of customers who have made purchases with the same salesperson.
    # **important** -> not able to solve (self - Join question)
SELECT t1.SalesPersonID,t1.CustomerID , t2.CustomerID
FROM demodatabase.sales1 t1
JOIN demodatabase.sales1 t2
ON t1.SalesPersonID = t2.SalesPersonID
AND t1.CustomerID != t2.CustomerID


```
