### Q9:- Given 2 fractions, find the sum of those 2 fractions.Take the numerator and denominator values of the fractions from the user.

````python
# Write your code here
import math

a1 = int(input('Enter numerator of 1st number: '))
b1 = int(input('Enter denominator of 1st number: '))
a2 = int(input('Enter numerator of 2nd number: '))
b2 = int(input('Enter denominator of 2nd number: '))

num = (a1 * b2) + (a2 * b1)
den = b1 * b2

gcd = math.gcd(num, den)

num = num // gcd
den = den // gcd

print('Sum =', num, '/', den)
````
