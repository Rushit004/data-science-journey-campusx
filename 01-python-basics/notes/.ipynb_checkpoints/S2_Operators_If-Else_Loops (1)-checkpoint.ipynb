1. Operators

   

```python
# Arithmatic operators
print(4+5)

print(10-5)

print(5*6)

print(5/2)

print(5//2) # 2.5 ->2

print(5%2)

print(5**2)
```

```python
# Relational operators
print(4>5)

print(4<5)

print(5>=4)

print(4<=5)

print(4==4)

print(4!=4)
```

```python
# Logical operator
print(1 and 0)

print(1 or 0)

print(not 1)

print(4 and 2)
```

```python
# Bitwise operators

print(2 & 3) # 010 & 011

print(2 | 3) # 010 | 011

print(2 ^ 3) # XOR 010 ^ 011

print(~3) # 011 ->100

print(4<<1)

print(4>>1)

```

```python
# Assignment operators

a =2
print(a)

a+=2
print(a)

a-=2
print(a)

a*=4
print(a)

a/=2
print(a)

# a++, a--,++a,--a not allowed
```

```python
# membership operator

# in/not in

print('D' in 'Delhi')

print('H' not in 'Hii')

print(1 in (2,5,2))
print('Rushit' in {'Name' : 'Rushit'})
print('Name' in {'Name':'Rushit'})# in dictionary it finds for key.
# works on all datatypes
```

```python
# find sum of 3 digit number entered by user

x =int(input( "Enter 3 digit number"))
print(x)

sum1=x%10
x=x//10

sum2=x%10
x=x//10

sum3=x%10
x=x//10

print(sum1+sum2+sum3)
```

2. if-elif-else



```python
# 1.login program 

# email->rt@gmail.com
# pass->123456

email = input("Enter Email")
password  =input("Enter Password")

if email=="rt@gmail.com" and password =='123456':
    print("Welcome user")
elif email=="rt@gmail.com" and password !='123456':
    print("Incorrect Password")
    password = input("Enter password again")

    if password =='123456':
        print("Welcome user")
    else :
        print("You cant do this")
    
elif email!="rt@gmail.com" and password =='123456':
    print("Incorrect Email")
    email = input("Enter Email again")

    if email =='rt@gmail.com':
        print("Welcome user")
    else :
        print("You cant do this")
else :
    print("Incorect Password or Email")
 
```

```python
# 2.find min of given 3 numbers
a=int(input("Enter 1st number:"))
b=int(input("Enter 2nd number:"))
c=int(input("Enter 3rd number:"))

if a<b and a<c:
   print('min =',a)
elif b<c:
    print("min =",b)
else :
    print("min =",c)
```

```python
# 3. Menu driven calculator

a = int(input('Enter 1st number:'))
b = int(input('Enter 2nd number:'))

op = input('Enter the operation (+,-,*,/)')

if op == '+':
    print(a+b)
elif op == '-':
    print(a-b)
elif op == '*':
    print(a*b)
elif op == '/':
    print(a/b)
else:
    print('invalid operator')



```

4. Loops in Python

   

```python
# While loop
n = int(input('Enter the number'))
i = 1
while i<11:
    print(n,' * ',i,' = ',n*i )
    i += 1
```

```python
# you can use else also in loop(in python , we can't use in c ,c++)

# While loop with else

x = 1
while x<3:
    print(x)
    x+=1
    
else :
    print('Limit crossed')
```

```python
# guss the number

import random
jackpot = random.randint(1,100)

x = int(input('Guess any number'))
cnt =1
while x!=jackpot:

   if x<jackpot:
     print('Try bigger number')
   else :
     print('Try smaller number')
   x = int(input('Enter  number again'))
   cnt+=1
else:
    print('Correct guess.')
    print('Attempts : ',cnt)
```

```python
# For loop 
for i in range(1,11,2):
    print(i)
```

```python
for i in range(10,0,-2):
    print(i)
```

```python
for i in 'Rushit':
    print(i)
```

```python
# One town has current population 10000 . which increases by 10% each year.Find population at each year   10 years from now  using loops 

curr_pop = 10000

for i in range(10,1,-1):
    print(i,curr_pop)
    i-=1
    curr_pop /=1.1  
```

```python
# sequence sum
# 1/1! + 2/2! + 3/3!
n = int(input('Enter number:'))

result = 0
fact =1
for i in range(1,1+n):
    fact = fact*i
    result += i/fact
print(result)
    
```
