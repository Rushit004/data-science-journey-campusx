2. *args and *kwargs
   
   This are special python keywords used to pass the variable length of arguments to function

```python
# *args
# allows us to pass a variable number of non-keyword argment to function

# def multiply(a,b):
#     return a*b
# multiply(2,3)

def multiply(*args): # internally it makes tuple # you can use any word in peramter e.g. (*l),(*Rushit) 
    product = 1
    for i in args:
        product = product*i
  # print(args)
    return product

print(multiply(1,2))
print(multiply(4,5,6))
print(multiply(4,5,6,10))
```

```python
# **kwargs
# **kwargs allows us to pass any number of keyword arguments
# keyword arguments mean that contains key value pair, like python dictionary

def display(**kwargs):# it converts arguments in dictionary # you can any word insted of kwargs
    for (key,value) in kwargs.items():
        print(key,'->',value)
display(india='Delhi',srilanka='colombo',nepal='katmandu')


# order of an argument matters(normal->*args->**kwargs)
# the words 'args' and 'kwargs' is are only convensional, you can use any other word instead of that.
```

7. Lambda Function

```python
# lambda function is small anonymous function
# lambda function can take any number of arguments, but can only have one expression
# syntax -> lambda a,b :a+b 
```

```python
# x -> x**2
a =lambda x : x**2
a(6)
```

```python
# x,y -> x+y
b = lambda x,y : x+y
b(4,5)
```

```python
# check if a string has 'a'
a = lambda s : 'a' in s
a('apple')
```

```python
# check odd or even
a = lambda x : 'even' if x%2==0 else 'odd'
a(4)
```

```python
# Diffrence between Lambda function and normal function

# no name
# lambda has no return value(infact,returns a function itself e.g. <function __main__.<lambda>(x)>)
# lambda is written in one line
# no reusability

# why to use lembda function ?
# -> they are used with higher order functions
```

8. Higher order function

```python
# def square(x):
#     return x**2

# # HOF
# def transform(f,l):
#     output =[]
#     for i in l:
#         output.append(f(i))
#     print(output)
    
# l = [1,2,3,4,5]
# transform(square,l)

# HOF 
def transform(f,l):
    output =[]
    for i in l:
        output.append(f(i))
    print(output)
    
l = [1,2,3,4,5]
transform(lambda x:x**2,l)
```

9. MAP
    
    It takes function and iterable as an perameter

```python
# Square the item of list

l = list(map(lambda x:x**2,[1,2,3,4,5]))
print(l)
```

```python
# odd/even labelling of list items
l = [1,2,3,4,5]
list(map(lambda x:'even' if x%2==0 else 'odd',l))
```

10. filter
