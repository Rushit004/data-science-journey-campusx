### Exceptions

If things go wrong during the execution of the program(runtime). It generally happens when something unforeseen has happened.

- Exceptions are raised by python runtime
- You have to takle is on the fly

#### **Examples**

- Memory overflow
- Divide by 0 -> logical error
- Database error

```python
# why it is important to handle exceptions(user experience , security)
# how to handle exceptions
# -> Try except block
```

```python
try:
  m=5
  f = open('sample1.txt','r')
  print(f.read())
  print(m)
  print(5/0)
  L = [1,2,3]
  L[10]
    
except FileNotFoundError:
  print('file not found')
except NameError:
  print('variable not defined')
except ZeroDivisionError:
  print("can't divide by 0")
except Exception as e:
  print(e)
```

```python
# else
try:
  f = open('sample.txt','r')
except FileNotFoundError:
  print('file not found.')
except Exception:
  print('Some thing issue')
else:
  print(f.read())
```

```python
# finally
# else
try:
  f = open('sample1.txt','r')
except FileNotFoundError:
  print('file nai mili')
except Exception:
  print('kuch to lafda hai')
else:
  print(f.read())
finally:
  print('This will always execute. ')
```

```python
class Bank:
    def __init__(self,balance):
        self.balance = balance
    def withdraw(self,amount):
        if amount <0:
            raise Exception('Amount can not be Negative.')
        if self.balance < amount:
            raise Exception('Bhai pehle apna balance bada.')
        self.balance = self.balance - amount

obj = Bank(10000)

try:
    obj.withdraw(-100)
except Exception as e:
    print(e)
else:
    print(obj.balance)
    
```

```python
# Coustom exception
\
class MyException(Exception):
    def __init__(self,message):
        print(message)
class Bank:
    def __init__(self,balance):
        self.balance = balance
    def withdraw(self,amount):
        if amount <0:
            raise MyException('Amount can not be Negative.')
        if self.balance < amount:
            raise MyException('Bhai pehle apna balance bada.')
        self.balance = self.balance - amount

obj = Bank(10000)

try:
    obj.withdraw(-100)
except MyException as e:
    pass
else:
    print(obj.balance)
    
```
