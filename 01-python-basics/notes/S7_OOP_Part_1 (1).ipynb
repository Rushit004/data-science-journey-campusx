```python
# ATM

class Atm:
    # cunstructor (it is special method that you don't need to call constructor.it automatically called when object is created  )
    def __init__(self):
        # print('I am  Constructor and i am  called.')
        self.pin=''
        self.balance = 0
        self.menu()
        

    def menu(self):
        user_input = input('''
        Hi how can i help you ?
        1. Create pin
        2. Change pin
        3. Check balance
        4. Withdraw
        5. Exit
        ''')

        if user_input == '1':
            # create pin
            self.create_pin()
            
        elif user_input == '2':
            # Change pin
            self.change_pin()
            
        elif user_input == '3':
            # chack balance
            self.check_balance()
            
        elif user_input == '4':
            # Withdraw
            self.withdrawl()

        else:
            exit()

    def create_pin(self):
        user_pin = input('Enter your pin:')
        self.pin = user_pin

        user_balance= int(input('Enter balance:'))
        self.balance = user_balance
        
        print('pin created sucessfully.')
        self.menu()

    def change_pin(self):
        old_pin=input('Enter old pin:')
        if old_pin == self.pin:
            # change pin
            new_pin=input('Enter new pin')
            self.pin=new_pin
            print('pin changed sucessfully.')
            self.menu()
        else:
            print('Wrong old pin.')
            self.menu()
            
            
    def check_balance(self):
       user_pin=input('Enter pin:')
       if user_pin==self.pin :
           print('balance =',self.balance)
       else :
            print('Wrong pin try again.')
            self.menu()

    def withdrawl(self):
        user_pin=input('Enter pin.')
        if user_pin == self.pin :
            withdrawl_amt=int(input('Enter amount to be withdrawl:'))
            if withdrawl_amt <= self.balance:
                self.balance-=withdrawl_amt
                print(withdrawl_amt , 'RS. Sucessfully withdrawl from your account and balance left is = ',self.balance)
                self.menu()
            else :
                print('Not valid amount in account.')
        else :
            print('Wrong pin try again.')
        self.menu()
            
            
        
        
        
    
```

```python
# create our own datatype
# make fraction datatype

class Fraction:

    def __init__(self,x,y): # parameterized constructor
        self.num = x
        self.den = y

    def __str__(self): # used to display some thing in specific syntax or way.
        return '{}/{}'.format(self.num,self.den)

    def __add__(self,other):
        new_num = self.num*other.den + other.num*self.den
        new_den = self.den*other.den

        return '{}/{}'.format(new_num,new_den)

    def __sub__(self,other):
        new_num = self.num*other.den - other.num*self.den
        new_den = self.den*other.den

        return '{}/{}'.format(new_num,new_den)

    def __mul__(self,other):
        new_num = self.num*other.num
        new_den = self.den*other.den

        return '{}/{}'.format(new_num,new_den)

    def __truediv__(self,other):
        new_num = self.num*other.den
        new_den = self.den*other.num

        return '{}/{}'.format(new_num,new_den)
    

    def convert_to_decimal(self):
        return self.num/self.den
    
    
        
        
                    


    
 
```

```python
f1 = Fraction(1,2)
print('f1 =',f1)
f2 = Fraction(4,5)
print('f2 =',f2)
f3 = f1 + f2
print('f1 + f2 =',f3)
f4 = f1 - f2
print('f1 - f2 =',f4)
f5 = f1 * f2
print('f1 * f2 =',f5)
f6 = f1 / f2
print('f1 / f2 =',f6)
f1.convert_to_decimal()
```
