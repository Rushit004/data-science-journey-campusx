```python
# create list

# empty list
print([])

# 1D list
print([1,2,3]) #homogeneous

# 2D list # hetrogeneous
print([1,2,3,[4,5,6]])

# 3D list
print([[1,2,1,2,3],[1,2,3,4]]) # homogeneous

# hetrogeneous list
print([1,4.5,True,'RT'])

# using typeconversion
print(list('Hello'))

```

```python
# Accessing items from list
l = [1,2,3,[4,5]]

# positive indexing
print(l[0])
print(l[2])
print(l[3][0])

# negetive indexing
print(l[-1])
print(l[-1][-2])

l =[  [[1,2],[3,4]],[[5,6],[7,8]]  ]
print(l)
print(l[0][1][0])



```

```python
# slicing
l = [[[1,2],[3,4]],[[5,6],[7,8]]]
print(l[0:1])
```

```python
# adding item to list
# append
l = [1,2,3,4]

l.append(5)
print(l)
```

```python
# extend
l = [1,2,3,4,5]
l.extend(['Rt',45]) # can append multiple arguments at a time.
print(l)
l.append([4,5,6]) # will appnend   single argument  at a time ([4,5,6]).
print(l)

l.extend('Rushit')
print(l)
```

```python
# insert 
l = [1,2,3,4,5,6]
l.insert(1,100) # l.insert(idx,value)
print(l)
```

```python
# Editing items in list
l = [1,2,3,4,5]
# editing with index
l[-1]=400
print(l)

# editing with slicing
l[1:4]=[200,300,400]
print(l)

```

```python
# Deletion in list

# del
# remove
# pop
# clear


# del (based on index position)
l = [1,2,3,4]

print(l)
# del l

del l[-1]
print(l)

del l[1:3]
print(l)

```

```python
# remove (based on value)
l = [5,1,2,3,5,4,5]

l.remove(5)
print(l)
```

```python
# pop(based on index)
l = [1,2,3,4,5]

l.pop(0)
print(l)

l.pop()# if idx not provided it deletes from last
print(l)
```

```python
# clear(makes list empty by deleting all data )

l = [1,2,3,4,5]

l.clear()
print(l)
```

```python
# arithmatic

l1 = [1,2,3,4]
l2 = [0,7,6,8]

print(l1+l2)# mearge

print(l1*3)
```

```python
# membership operator


l1 = [1,2,3,4]
l2 = [2,7,6,[5,6]]
print(5 in l1)
print(5 not in l2)
print([5,6] in l2)



```

```python
# loops

l = [[1,2,[3,4],['rt',18]],['Rushit','tholiya']]
for i in l:
    print(i)

```

```python
# list functions
# 1.) common on all datatypes
# len()
# min()
# max()
# sorted()

l = [[1,2,3,4],[5,6]]
print(len(l))
print(min(l))
print(max(l))
print(sorted(l,reverse=True))
```

```python
# 2.)
# Count()
l = [1,2,3,4,1,2]
print(l.count(1))
# index
print(l.index(2))
#reverse()
print(l.reverse())# it permently changes (reverse the list0)
print(l)
# sorted()
print(sorted(l))
print(l)
# sort()
print(l.sort())# perment changes(unlike sorted())
print(l)

# copy()
l = [1,2,3]
print(l)
print(id(l))
l1 = l.copy()
print(l1)
print(id(l1))
```

```python
# list comprehension
# it provides concise way of creating list

# newlist = [expresion for item in iterable if condition == True ]
# more time-efficient and space efficient than loop
# require fewer lines of code
# transform iterative statment in formula


```

```python
# add 1 to 10 in list
l=[]
for i in range(1,11):
    l.append(i)
print(l)

# list comprehension
l = [i for i in range(1,11)]
print(l)
```

```python
# scaler multiplication on vector
v = [1,2,3]
s = -2

x = [s*i for i in v ]
print(x)
```

```python
# add squeres
l = [1,2,3,4,5,6]

s = [i**2 for i in l]
print(s)
```

```python
# print all numbers divisible by 5 in range of 1 to 50
l = [i for i in range(1,51) if i%5==0]
print(l)
```

```python
# find programming language that starts whith latter 'p'.
languages = ['python','java','php','c','cpp']
# [i for i in languages if i[0]=='p']
[language for  language in languages if language.startswith('p') ]
```

```python
# nested if with list comprehension

basket = ['apple','guava','cherry','banana']
my_fruits = ['apple','kiwi','garpes','banana']

# add mew list from my_fruits and items if the fruit exist in basket and also starts with 'a'

[fruit for fruit in my_fruits if fruit in basket if fruit.startswith('a')]
```

```python
# print 3*3 matrix
[[i*j for i in range(1,11)] for j in range  (1,11) ]
```

```python
# cartesian product
l1 = [1,2,3]
l2 = [4,5,6]
[i*j for i in l1  for j in l2]
```

2. 2 ways to traverse a list
   
   itemwise
   
   indexwise

```python
# itemwise
l = [1,2,3,4,5]
for i in l:
    print(i)
```

```python
# indexwise
l = [1,2,3,4,5]

for i in range(0,len(l)):
    print(l[i])
```

3. Zip function


```python
# it returns Zip object,which is an iterator of tuples where thw first item in each passed iterator is paired together, and then the  second item in each passed iterator are paired together
# if the passed iterator has diffrent length, the iterator with least item decide  length of new iterator


# WAP to multiply items of 2 list indexwise
l1 = [1,2,3,4]
l2 = [-1,-2,-3,-4,-5]
print(list(zip(l1,l2)))


[i*j for i,j in zip(l1,l2)]
```
