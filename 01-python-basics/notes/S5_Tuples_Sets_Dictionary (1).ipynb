1. Tuples(similar to list but it is imutable)
   
   ordered
   
   unchangable
   
   allows duplicate
   

```python
# creating tuple

#empty
t1 = ()
print(t1)
t2 = (2)
print(t2)
print(type(t2))
t2 = (2,)# single item tuple
print(t2)
print(type(t2))

# homogeneous
t3 = (1,2,3,4)
print(t3)

# hetrogeneous tuple

t4 = (1,20.2,True,'RT',('Rushit','Tholiya'))
print(t4)
# using type conversion
t5 = tuple('Rushit')
print(t5)
```

```python
# Accessing items from tuple

# index
# slicing
print(t3)
print(t3[0])
print(t3[-1])

# slicing
print(t3[0:2])
print(t3[0:4:2])
print(t3[-1:-5:-2])
```

```python
# How to edit items

t1 = (1,2,3,4,5)
 #t3[0]=1 Error(immutable)
```

```python
# adding items
print(t3)
# not possible(like insert(),append() etc.)
```

```python
# deleting item
t3 = (1,2,3,4)
print(t3)
del t3
# print(t3) Complete tuple can be deleted but
# print(del t[0]) portion can't be deleted
```

```python
# operations on tuples
# + and *

t1 = (1,2,3)
t2 = (2,3,4)

print(t1+t2)
print(t1*3)
# membership
print(1 in t1)
# loop
for i in t1:
    print(i)
```

```python
# tuple functions
# len
# min
# max
# sum
# sorted

t = (1,3,0,40,5)
print(len(t))
print(min(t))
print(max(t))
print(sum(t))
print(sorted(t,reverse =True))
```

```python
# count
t = (2,2,1,1,1,1,5,4)
print(t.count(1))# gives count of value

# index
print(t.index(5))# gives  index  of value 
```

```python
# diffrence between list and tuple

# syntax
# mutability
# speed(tuples are slightly  faster then list)
import time
l = list(range(100000000))
t = tuple(range(100000000))

start = time.time()


for i in l:
    i*5
print('Time taken by list = ',time.time()-start)
start1 = time.time()
for i in t:
    i*5
print('Time taken by tuple = ',time.time()-start1)


# memory(list>tuple)
import sys
l = list(range(10000000))
t = tuple(range(10000000))

print('Memory space occupied by list = ',sys.getsizeof(l))
print('Memory space occupied by tuple = ',sys.getsizeof(t))

# builtin function(list>tuple)
# error prone(list>tuple)
# usability 
```

```python
# tuple unpacking
a,b,c =(1,2,3)
print(a,b,c)
# a,b = (1,2,3) # not possible  

# swap in python
print(a,b)
a,b = b,a
print(a,b)

#store othervalues
a,b,*others = (1,2,3,4)
print(a,b)
print(others)

a,b,*x= (1,2,[4,5,6],'RT',('Heloo'))
print(a,b)
print(x)
```

```python
# ziping tuple
a =(1,2,3,4)
b = (5,6,7)
list(zip(a,b))
tuple(zip(a,b))

```

2. Sets
   
   unorderd
   
   mutable
   
   no duplicates
   
   can't contain mutable data types


```python
# creating set
# empty
s ={}
print(type(s))
s = set()
print(s)
print(type(s))
# 1D 
s1 = {1,1,2,3}
print(s1)
# 2D (not possible)
# s2 = {1,2,3,{4,5}} # Error because it can't contain mutable datatype and Set is mutable data type so 2D set can't be possible
# hetrogeneous set
s3 = {1,'Rt',4.5,True}
print(s3)
s4 = { 1,(1,2),2}
print(s4)

# using typeconversion
s5 = set([1,2,4,5,1])
print(s5)
# set can't have mutable items
# s6 = {1,2,[4,5]}
# print(s6)

# set is unorderd
s1 = {1,2,3}
s2 = {3,1,2}
print(s1==s2)
```

```python
# Accessing items
# indexing method not allowed because of set is unorderd
# s ={1,2,3}
# print(s[0])


```

```python
# Editing items
# s1 = {1,2,3}
# s1[0]=100 not allowed because  set is unorderd
```

```python
# adding items
# add
s = {1,2,3,4}
s.add('RT') # position where item will be added depends on hashing
print(s)
# update
s.update([4,5,6]) # add multiple items
print(s)
```

```python
# deleting items


# del
s = {1,2,3,4,5}
del s
# del s[0] # cant delete perticular portion 
#print(s)
s = {1,2,3,4,5}
# discard(based on value)
s.discard(6)# dont throw error if value does't exist
print(s)
# remove(based on value)
s.remove(5)# throws error if value does't exist
print(s)

# pop
s.pop()# randomly deletes any item
print(s)
# clear
s.clear() #delete all values
print(s)
```

3. set operations
   
   union(|)
   
   intersection(&)
   
   diffrence(-)
   
   symmetric diffrence(^)
   
   membership test
   
   iteration
   

```python
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

# union
print(s1 | s2)
# intersection
print(s1 & s2)
# diffrence
print(s1 - s2)
# summetric diffrence
print(s1 ^ s2)
# membership test
print(1 in s1)
# iteration
for i in s1:
    print(i)
```

```python
# set functions
# len
# min
# max
# sum
# sorted

s = {1,2,3,4,5,6}
print(len(s))
print(max(s))
print(min(s))
print(sum(s))
print(sorted(s,reverse=True))# always comes in list
```

```python
# union/union_update


s1 = {1,2,3,4,5}
s2 = {3,4,5,6,7}
# same as (|)
print(s1.union(s2))
print(s1.update(s2))# permently change s1  by s1 union s2
print(s1)
print(s2)#s2 as it is
```

```python
# intersection/intersection_update

s1 = {1,2,3,4,5}
s2 = {3,4,5,6,7}
# same as (&)
print(s1.intersection(s2))
print(s1.intersection_update(s2))# permently change s1  by s2 intersection s2
print(s1)
print(s2)#s2 as it is
```

```python
# diffrence/diffrence_update
s1 = {1,2,3,4,5}
s2 = {3,4,5,6,7}
# same as (-)
print(s1.difference(s2))
print(s1.difference_update(s2))# permently change s1  by s1 diffrence s2
print(s1)
print(s2)#s2 as it is
```

```python
# symmetric_diffrence/symmatric_diffrence_update
s1 = {1,2,3,4,5}
s2 = {3,4,5,6,7}
# same as (^)
print(s1.symmetric_difference(s2))
print(s1.symmetric_difference_update(s2))# permently change s1  by s1 symmetric_diffrence s2
print(s1)
print(s2)#s2 as it is
```

```python
# isdisjoint
s1 = {1,2,3,4,5,6}
s2 ={1,2,3,6}
print(s1.isdisjoint(s2)) # disjoint sets are those in which there is no common item from both sets
# issubset
print(s1.issubset(s2))
# issuperset
print(s1.issuperset(s2))
```

```python
# copy
s1 = {1,2,3,4}
s2 = s1.copy()
print(s1,s2)
```

```python
# Frozen set
# it is just an immutable version of python set object

# creating frozen set

fs = frozenset([1,2,3,4])
print(fs)
```

```python
# what works and what does not
fs1 = frozenset({1,2,3,4})
fs2 = frozenset({0,8,9,4})

print(fs1 | fs2) # all function that does not make changes permenenty
# works -> all read function
# doesn't work -> all write functions

# 2d frozenset possible? (yes)
fs = frozenset([1,2,frozenset([4,5])])
print(fs)


```

4. Dictionary
   
   mutable
   
   index has no meaning
   
   keys's can't be duplicate
   
   key's can't be mutable items
   
   unorderd
   

```python
# create Dictionary

# empty dictionary
d = {}
print(d)
# 1D dictionary
d1 = {'name':'Rushit','age':18}
print(d1)
# with mixed keys

d2 = {'name':'Rushit',(1,2,3):1}
print(d2)
# 2D dictionary
d3 = {
    'name':'Rushit',
     'collage':'Nirma university',
    'subjects':{
        'maths':89,
        'OOP':95,
        'DSA': 85
    }
}
print(d3)
print(d3['subjects']['maths'])
# using sequence and dict function
d4 = dict([('name','Rushit'),('age',18)])
print(d4)

# you can't have duplicate keys
d5 = {'name':'Rushit','name':'RT'}# always considerd last duplcate key 
print(d5)

# mutable items as keys 
# d6 = {'name':'Rushit',[1,2,3]:2}
```

```python
# Accessing  items
d = {'name':'Rushit','age':18}
print(d)
# print(d[0]) not allowed

# using key
print(d['name'])
# get function
print(d.get('age'))
```

```python
# Adding new key value pair
d = {'name':'Rushit','age':18}
d['rollno'] = '24BCE333'
print(d)
```

```python
# remove key value pair
d = { 'name': 'Rushit', 'age': 18, 'rollno': '24BCE333','cgpa':8.55,'sem':4,1:'RT'}
print(d)

# pop   syntax-> # d.pop(key)
d.pop(1)
d.pop('cgpa')
print(d)

# popitem #pop last item in dictionary
d.popitem()
print(d)

d1 = {'name':'RT','age':18}
# del
del d1['name']
print(d1)
del d1
# print(d1)

# clear
# d1.clear()
# print(d1)
```

```python
# editing key value pairs

print(d3)
d3['subjects']['OOP'] = 98
print(d3)
```

```python
# operations in dictionary
# membership

print(d3)
# print('Nirma university' in d3)-> False # it finds keys not values
print('name' in d3)

# iteration
d = {'name' : 'Rushit','age' : 18,'cgpa':8.55}
for i in d:
    print(i)

for i in d:
    print(i,d[i],sep=' -> ')
```

```python
# dictionary functions
# len
# sorted
# min (based on ASCII)
# max (based on ASCII)

print(d)

print(len(d))
print(sorted(d,reverse = True))
```

```python
# items
# keys
# values

print(d.items())
print(d.keys())
print(d.values())
```

```python
# update
d1 = {1:1,2:4,3:9}
d2 = {4:16,5:25,6:36}

print(d1.update(d2))
print(d1)
print(d2)
```

```python
# dictionary comprehension
# {key: value for vars in iterable}

# print 1st 10 numbers and their squers
{i:i**2 for i in range(1,11)}
```

```python
# using existing dictionary
# convert distance to mile from km
distances = {'Delhi':1000,'mumbai':2000,'bangalore':3000}
{key:value*0.621 for (key,value) in distances.items() }
```

```python
# using Zip
days = ['mon','tue','wed','thu','fri','sat','sun']
temp = [30.1,30.5,34.3,36.5,32.8,30.9,39.0]
{key : value for (key,value) in zip(days,temp)}
```

```python
# using if condition
product ={'laptop':45,'ipad':0,'tablet':23}
{key:value for (key,value) in product.items() if value >0}
```

```python
# nested comprehension
# print table of number from 2 to 4

{i:{j:i*j for j in range(1,11)} for i in range(2,5)}
```
