1. Nested loops

```python
# Unique pairs

n = int(input('Enter the number'))

for i in range(1,1+n):
    for j in range(1,1+n):
        print(i , j,sep=',' )
```

```python
#1.)
n=int(input('Enter no. of rows'))

for i in range(1,1+n):
    for j in range(1,i+1):
        print('*',end ='')
    print()
```

```python
#2.)
n = int(input('Enter number'))

for i in range(1,1+n):
    for j in range(1,1+i):
        print(j,end='')
    for k in range(i-1,0,-1):
        print(k,end='')
    print()
```

2. Loop control statements
   
   Break
   
   Continue
   
   Pass
   

```python
# Break
for i in range(1,10):
    if i==5:
        break
    print(i)


```

```python
# find prime numbers
a = int(input('Enter lower range'))
b = int(input('Enter Upper range'))

for i in range(a,b+1):
    for j in range(2,i):
        if i%j == 0 :
            break
    else : 
        print('Prime no :',i)
```

```python
# Continue

for i in range(1,11):
    if i == 5:
        continue
    print(i)
```

```python
# Pass Statement(used when you dont want to write any logic in that)

for i in range(1,11):
    pass
    
```

```python
# Create String
s = 'Rushit'
s = "Rushit"
s = '''Rushit'''# multi line strings
s = """Rushit"""
# using str() function
s = str('12')
```

```python
# Indexing and Slicing
s = 'Rushit Tholiya' # index starts from 0 and if Negetive indexing it starts from -1
print(s[0])
print(s[0:4])
print(s[4:6])

print(s[-1])
print(s[-7:])

print(s[0:7:2])
print(s[5::-1])
print(s[::-1])

print(s[-1:-8:-1])
```

```python
# operation on Strings

# Arithmetic
print('Rushit' +' '+ 'Tholiya')
print(('Rushit' + ' ')*3)
print('*'*5)

# Relational

print('Rushit'=='Tholiya')
print('abcd' > 'abc') # Lexicographically comparision
print('AB'>'ab') # ASCII of A < a 

# Logical

print('Hello' and 'World') 
print('Hello' or 'World') 
print('' and 'World') # if string is empty then False
print(''or 'World')
print(not '')
print(not 'Hii')
```

```python
# String Functions
# 1.)Works on all datatypes
# len
# max
# min
# sorted

print(len('Rushit Tholiya'))
print(max('Rushit Tholiya'))
print(min('Rushit Tholiya'))
print(sorted('Rushit Tholiya'))
```

```python
# 2.) Only applicable to strings
# Capitalize()
# Title()
# upper()
# lower()
# swapcase()

s = 'rushit tholiya'
print(s.capitalize())
print(s.title())
print(s.upper())
print(s.lower())

s1 = 'rUShit ThoLIyA'
print(s1.swapcase())
```

```python
# 3.)
# count()
# find()
# index()
print('My name is Rushit'.count('i'))
print('My self Rushit Tholiya'.find('rush')) # if doesn't exits give -1
print('My self Rushit Tholiya'.index('Rush')) # if doesn't exist gives Error
```

```python
#4.)
# endswith()
# strartswith()
print('Rushit Tholiya'.endswith('ya'))
print('Rushit Tholiya'.startswith('ru'))
```

```python
# 5.)
# format()

name ='Rushit'
age  = 18
print('Hii i am {} and i am {} year old.'.format(name,age))
print('Hii i am {1} and i am {0} year old.'.format(name,age))
```

```python
# 6.)
# isalnum
# isalpha
# isdigit
# isidentifier

print('rushit004'.isalnum())# is alph 'or' is numeric #(not 'and')
print('rushit'.isalnum())
print('004'.isalnum())
print('123'.isalpha())
print('RT004'.isdigit())
print('_004'.isidentifier())
print('_$'.isidentifier())
```

```python
# 7.)
# split
# join 

print('Hii Iam Rushit'.split()) # default ' '
print('Hii Iam Rushit'.split('i')) 
print(' '.join(['Hii','Iam','Rushit','Tholiya']))
print('/'.join(['Hii','Iam','Rushit','Tholiya']))
```

```python
# count frequency of character in string
# e.g. 'Hello World' , l ->3

s = input('Enter the string')

x =input('Enter character whoes frequency needs to find.')
# print(s.count(x))
cnt=0
for i in s:
    if i == x:
        cnt+=1
print('Frequency of ',x, '=',cnt)
```

```python
# WAP to remove perticular character from given string

s = input('Enter the string:')

x =input('Enter character to be removed.:')

for i in s:
    if i==x:
        continue
    print(i,end ='')
```

```python
# WAP to find string is  palindrome or not

s = input('Enter the string:')
flag =True
for i in range(0,len(s)//2):
    if s[i] != s[len(s)-i-1]:
        print('String is not an Palindrome')
        flag = False 
        break 
if flag:
    print('String is Palindrome')
        
```

```python
# WAP to count number of words in string without using split()

s = input('Enter the string:')
cnt=0
for i in s:
    if i == ' ':
      cnt+=1
print('Total number of words =',cnt+1)
```

```python
# WAP to store  words of string in list without using split()
s = input('Enter the string:')
l=[]
temp =''
for i in s:
    if i  !=' ':
      temp+=i
    else :
        l.append(temp)
        temp =''
l.append(temp)
print(l)
```

```python
# WAP to convert string to title case without using title()

s = input('Enter the string:')
l = []
for i in s.split():
    l.append(i[0].upper()+i[1:].lower())

print(' '.join(l))
```

```python
# convert given integer to string

a = int(input('Enter number:'))

digits = '0123456789'
result =''
while a!=0:
    result = digits[a%10]+result
    a//=10
print(result)
```
