```python
# agg method
genres.agg({
    'Runtime':'mean',
    'IMDB_Rating':'mean',
    'No_of_Votes':'sum',
    'Gross':'sum',
    'Metascore':'min'
})
```

```python
# passing list

genres.agg(['sum','min'])
```

```python
# Adding both the syntax
genres.agg(
    {
        'Runtime':['min','mean'],
        'IMDB_Rating':'mean',
        'No_of_Votes':['sum','max'],
        'Gross':'sum',
        'Metascore':'min'
    }
)
```

```python
# looping on groups
# find the highest rated movie of each genre

data_to_append = []
for group,data in genres:
    new_df = data[data['IMDB_Rating'] == data['IMDB_Rating'].max()]
    print(new_df)

```

```python
# split (apply) combine
# apply -> builtin function
genres.apply(min)
```

```python
# find number of movies starting with A for each group
def foo(group):
  return group['Series_Title'].str.startswith('A').sum()
  
```
